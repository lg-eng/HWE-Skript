## Aufbau eines Projektes

### Ziele
Hier werden die Ziele formulier welche erreicht werden sollen. Als Qualitätsmerkmal eines Ziel ist die Überprüfung auf SMART
* S ... Spezifisch
* M ... Messbar
* A ... Erreichbar (achievable)
* R ... Realistisch
* T ... Terminiert (Es gibt einen Anfang und ein Ende)

### Ist - Situation
Bevor überlegt werden kann wie die Ziele erreicht werden können muss überlegt werden welche Mittel zur verfügung stehen. Unter Mittel ist jede Form von Resource gemeint, die hilfreich sein kann das Ziel zu erreichen. Wissen, Hardware, Software, Mitarbeiter*innen.
